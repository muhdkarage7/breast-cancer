```python
import os
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, TrainingArguments, Trainer
from peft import PeftModel, LoraConfig
from datasets import Dataset, DatasetDict
import pandas as pd

model_name = "t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_name, token='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')
base_model = AutoModelForSeq2SeqLM.from_pretrained(model_name, token='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')

lora_config = LoraConfig(
    task_type="SEQ_2_SEQ_LM",
    r=8,
    lora_alpha=16,
    lora_dropout=0.1
)

model = PeftModel(base_model, lora_config)

top_p = 0.9
top_k = 150
temperature = 0.7

generation_args = {
    "do_sample": True,
    "top_k": top_k,
    "top_p": top_p,
    "temperature": temperature,
    "max_length": 100
}
```

```python
from sklearn.model_selection import train_test_split
from datasets import Dataset, DatasetDict
import pandas as pd

def preprocess_data(data, tokenizer):
    inputs = tokenizer(data["Question"].tolist(), padding="max_length", truncation=True, return_tensors="pt")
    labels = tokenizer(data["Answer"].tolist(), padding="max_length", truncation=True, return_tensors="pt")["input_ids"]

    inputs["labels"] = labels
    inputs["decoder_input_ids"] = labels
    inputs["decoder_attention_mask"] = tokenizer(data["Answer"].tolist(), padding="max_length", truncation=True, return_tensors="pt")["attention_mask"]

    return Dataset.from_dict(inputs)

X = dataset["Question"].tolist()
Y = dataset["Answer"].tolist()
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

train_data = pd.DataFrame({"Question": X_train, "Answer": Y_train})
test_data = pd.DataFrame({"Question": X_test, "Answer": Y_test})

train_dataset = preprocess_data(train_data, tokenizer)
test_dataset = preprocess_data(test_data, tokenizer)
```

```python
training_args = TrainingArguments(
    output_dir="./peft_model",
    per_device_train_batch_size=8,
    num_train_epochs=3,
    logging_dir="./logs",
    learning_rate=5e-5,
    weight_decay=0.01,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)
```

```python
trainer.train()
```
